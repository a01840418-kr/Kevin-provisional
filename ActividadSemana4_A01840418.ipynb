* 1a) La métrica G_mean se recomienda en problemas de clasificación con clases desbalanceadas, donde una clase tiene muchos más datos que otra. Su objetivo es medir qué tan bien el modelo identifica correctamente tanto la clase positiva como la negativa, evitando que el modelo se sesgue hacia la clase mayoritaria.

* 1b) La diferencia principal es que la G_mean evalúa el equilibrio entre sensibilidad y especificidad, mientras que la F1-score combina precisión y recall de la clase positiva. La ventaja de G_mean es que considera el desempeño en ambas clases, aunque no toma en cuenta directamente la precisión. Por otro lado, F1-score es útil cuando interesa principalmente detectar correctamente la clase positiva, pero puede ignorar el desempeño sobre la clase negativa.

* 1c) En el problema del derrame de petróleo no es conveniente usar únicamente la accuracy, porque los datos están desbalanceados y un modelo podría obtener alta exactitud prediciendo casi siempre “no derrame”. Esto sería engañoso, ya que no detectaría los derrames reales. La G_mean es más adecuada porque evalúa el equilibrio entre detectar correctamente los derrames y clasificar bien los casos sin derrame.

* 1d) Además de la G_mean, sería útil considerar métricas como Recall, para medir cuántos derrames reales se detectan; Precision, para evaluar cuántas detecciones son correctas; F1-score, para equilibrar precisión y recall; y ROC-AUC o PR-AUC, que ayudan a evaluar la capacidad general del modelo en problemas con clases desbalanceadas.

### +++++++++++ Inicia sección para agregar texto ++++++++++++++++

* 2a) La transformación Yeo-Johnson es una técnica de transformación de datos utilizada para hacer que una variable tenga una distribución más cercana a la normal. Ayuda a reducir la asimetría y mejorar el desempeño de algunos modelos de aprendizaje automático y métodos estadísticos. Su principal ventaja es que puede aplicarse tanto a valores positivos como negativos y cero.

* 2b) La principal diferencia entre la transformación Yeo-Johnson y la Box-Cox es que Box-Cox solo puede aplicarse a datos estrictamente positivos, mientras que Yeo-Johnson admite valores positivos, negativos y cero. Por ello, Yeo-Johnson es más flexible y útil cuando el conjunto de datos contiene números negativos o ceros



### +++++++++++ Termina sección para agregar texto ++++++++++++++

```python
# Transformaciones numéricas evitando el filtrado de información:

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer

numericas_pipeline = Pipeline(steps = [('impMediana', SimpleImputer(strategy='mean')),
                                       ('yeo-johnson', PowerTransformer(method='yeo-johnson')),
                                       ('std_transformer', StandardScaler()),
                                       ])

numericas_pipeline_nombres = Xtrain.columns

columnasTransformer = ColumnTransformer(transformers = [('num', numericas_pipeline, numericas_pipeline_nombres)],
                                        remainder='passthrough')
```

```python
from sklearn.metrics import make_scorer
from imblearn.metrics import geometric_mean_score

# Definamos la siguiente variable con la métrica G-mean a utilizar en los modelos:

mi_metrica = make_scorer(geometric_mean_score)
```
